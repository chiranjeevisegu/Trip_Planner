```python
from langchain.schema import Document
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
import json

# 1. Load the JSON data
with open('places.json', 'r') as f:
    places_data = json.load(f)

# 2. Convert each place into a Document object
documents = []
for place in places_data:
    # Create metadata dictionary (all fields except description)
    metadata = {
        "name": place["name"],
        "city": place["city"],
        "type": place["type"],
        "entry_fee": place["entry_fee"],
        "duration_hours": place["duration_hours"],
        "rating": place["rating"]
    }

    # Create Document with description as page_content and other fields as metadata
    doc = Document(
        page_content=place["description"],
        metadata=metadata
    )
    documents.append(doc)

# 3. Initialize embeddings model
# Using HuggingFace's all-MiniLM-L6-v2 model (small but effective)
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# 4. Create and save the FAISS vector store
vector_db = FAISS.from_documents(documents, embeddings)
vector_db.save_local("places_faiss_index")

print("Vector database created and saved successfully!")
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.llms import HuggingFacePipeline
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM
import json
import os

# Verify FAISS index exists
if not os.path.exists("/content/places_faiss_index"):
    raise FileNotFoundError("FAISS index not found at /content/places_faiss_index")

# 1. Load the FAISS vector store
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vector_db = FAISS.load_local("/content/places_faiss_index", embeddings, allow_dangerous_deserialization=True)
print("FAISS index loaded successfully")

# 2. Initialize the LLM
model_name = "google/flan-t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
llm_pipeline = pipeline(
    "text2text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length=256,
    min_length=50,
    batch_size=1,
    device=-1  # Use GPU with device=0 if available
)
llm = HuggingFacePipeline(pipeline=llm_pipeline)
print("LLM initialized")

# 3. Define the prompt template
prompt_template = """
You are a travel planner. Create a detailed travel itinerary for {city} based on the user's preferences, budget, and days. Include place names, entry fees, and duration, fitting within the budget and 8 hours/day.

**User Preferences**: {user_query}
**Budget**: {budget} INR
**Days**: {days}
**Places**:
{places_info}

**Instructions**:
- Select relevant places matching the preferences.
- Create a day-by-day itinerary.
- Total entry fees must be within {budget} INR.
- Total duration per day must not exceed 8 hours.
- Provide a brief justification.
- If no places fit, suggest alternatives.

**Output Format**:
**Travel Itinerary for {city}**
**Budget**: {budget} INR
**Days**: {days}

**Day 1**:
- Place: [Name], Entry Fee: [Fee] INR, Duration: [Hours] hours
...

**Justification**:
[Why these places were chosen.]

**Total Entry Fees**: [Sum] INR
"""
prompt = PromptTemplate(
    input_variables=["city", "user_query", "budget", "days", "places_info"],
    template=prompt_template
)

# 4. Create the RAG pipeline
llm_chain = LLMChain(llm=llm, prompt=prompt)
print("RAG pipeline created")

# 5. Function to generate itinerary
def generate_itinerary(city, user_query, budget, days):
    print("Starting itinerary generation...")
    # Retrieve relevant places
    print("Performing similarity search...")
    similar_docs = vector_db.similarity_search(user_query, k=10)
    print(f"Retrieved {len(similar_docs)} places")

    # Filter by city
    filtered_docs = [doc for doc in similar_docs if doc.metadata['city'].lower() == city.lower()]
    print(f"Filtered to {len(filtered_docs)} places in {city}")

    # Format places for prompt
    places_info = ""
    for doc in filtered_docs[:5]:  # Limit to 5 places
        places_info += (f"- Name: {doc.metadata['name']}, "
                        f"City: {doc.metadata['city']}, "
                        f"Type: {doc.metadata['type']}, "
                        f"Entry Fee: {doc.metadata['entry_fee']} INR, "
                        f"Duration: {doc.metadata['duration_hours']} hours, "
                        f"Rating: {doc.metadata['rating']}, "
                        f"Description: {doc.page_content}\n")
    print("Formatted places_info for prompt")

    # Run the LLM chain
    print("Running LLM chain...")
    response = llm_chain.run(
        city=city,
        user_query=user_query,
        budget=budget,
        days=days,
        places_info=places_info
    )
    print("LLM chain completed")

    # Save output
    with open("/content/itinerary.txt", "w") as f:
        f.write(response)
    print("Itinerary saved to /content/itinerary.txt")

    return response

# 6. Example usage
city = "Ooty"
user_query = "I want to see waterfalls and nature spots"
budget = 200  # in INR
days = 2

itinerary = generate_itinerary(city, user_query, budget, days)
print(itinerary)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.llms import HuggingFacePipeline
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM
import json
import os
import re

# Verify FAISS index exists
if not os.path.exists("/content/places_faiss_index"):
    raise FileNotFoundError("FAISS index not found at /content/places_faiss_index")

# 1. Load the FAISS vector store
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vector_db = FAISS.load_local("/content/places_faiss_index", embeddings, allow_dangerous_deserialization=True)
print("FAISS index loaded successfully")

# 2. Initialize the LLM
model_name = "google/flan-t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
llm_pipeline = pipeline(
    "text2text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length=512,  # Increased for better output
    min_length=50,
    batch_size=1,
    do_sample=False,  # Deterministic output
    device=-1  # Use GPU with device=0 if available
)
llm = HuggingFacePipeline(pipeline=llm_pipeline)
print("Device set to use CPU" if llm_pipeline.device == -1 else "Device set to use GPU")
print("LLM initialized")

# 3. Define the prompt template (simplified)
prompt_template = """
You are a travel planner. Create a travel itinerary for {city} based on the user's preferences, budget, and days. Include place names, entry fees, and duration, fitting within the budget and 8 hours/day. Reserve ₹200/day for lunch.

**Preferences**: {user_query}
**Budget**: {budget} INR (includes lunch)
**Days**: {days}
**Places**:
{places_info}

**Instructions**:
- Select up to 2 places per day matching the preferences.
- Include lunch (₹200/day).
- Total cost (entry fees + lunch) must not exceed {budget} INR.
- Total duration per day must not exceed 8 hours.
- Provide a brief justification.

**Output Format**:
**Itinerary for {city}**
Day 1:
- Place: [Name], Entry Fee: [Fee] INR, Duration: [Hours] hours
- Lunch, Cost: 200 INR
...
Justification: [Why these places were chosen.]
Total Cost: [Sum] INR
"""
prompt = PromptTemplate(
    input_variables=["city", "user_query", "budget", "days", "places_info"],
    template=prompt_template
)

# 4. Create the RAG pipeline
llm_chain = LLMChain(llm=llm, prompt=prompt)
print("RAG pipeline created")

# 5. Function to generate itinerary
def generate_itinerary(city, user_query, budget, days):
    print("Starting itinerary generation...")
    # Calculate budget after lunch costs
    lunch_cost = 200 * days
    available_budget = budget - lunch_cost
    if available_budget < 0:
        return "Error: Budget is too low to cover lunch costs (₹200/day)."

    # Retrieve relevant places
    print("Performing similarity search...")
    similar_docs = vector_db.similarity_search(user_query, k=10)
    print(f"Retrieved {len(similar_docs)} places")

    # Filter by city and budget
    filtered_docs = [
        doc for doc in similar_docs
        if doc.metadata['city'].lower() == city.lower()
        and doc.metadata['entry_fee'] <= available_budget
    ]
    print(f"Filtered to {len(filtered_docs)} places in {city}")

    # Format places for prompt
    places_info = ""
    for doc in filtered_docs[:5]:  # Limit to 5 places
        places_info += (f"- Name: {doc.metadata['name']}, "
                        f"Entry Fee: {doc.metadata['entry_fee']} INR, "
                        f"Duration: {doc.metadata['duration_hours']} hours, "
                        f"Description: {doc.page_content}\n")
    print("Formatted places_info for prompt")

    # Run the LLM chain
    print("Running LLM chain...")
    response = llm_chain.run(
        city=city,
        user_query=user_query,
        budget=budget,
        days=days,
        places_info=places_info
    )
    print("LLM chain completed")
    print("Raw LLM response:\n", response)  # Debug raw output

    # Format output with fallback
    formatted_itinerary = format_output(response, city, budget, days, lunch_cost, filtered_docs)

    # Save output
    with open("/content/itinerary.txt", "w") as f:
        f.write(formatted_itinerary)
    print("Itinerary saved to /content/itinerary.txt")

    return formatted_itinerary

# 6. Function to format LLM output with fallback
def format_output(response, city, budget, days, lunch_cost, filtered_docs):
    formatted = f"🗓️ Travel Itinerary for {city}\n"
    formatted += f"💰 Budget: ₹{budget}\n"
    formatted += f"📅 Days: {days}\n\n"

    lines = response.split("\n")
    total_cost = lunch_cost
    day = 0
    day_hours = [0] * days
    places_added = [[] for _ in range(days)]

    # Parse LLM output
    for line in lines:
        line = line.strip()
        if line.startswith("Day"):
            day = min(int(re.search(r'\d+', line).group()), days)
            formatted += f"🗓️ Day {day}:\n"
        elif line.startswith("- Place:"):
            match = re.match(r"- Place: (.*?),\s*Entry Fee: (\d+) INR,\s*Duration: ([\d.]+) hours", line, re.IGNORECASE)
            if match and day <= days:
                name, fee, duration = match.groups()
                fee = int(fee)
                duration = float(duration)
                if total_cost + fee <= budget and day_hours[day-1] + duration <= 8:
                    formatted += f"- {name} (₹{fee}, {duration} hours)\n"
                    total_cost += fee
                    day_hours[day-1] += duration
                    places_added[day-1].append(name)
        elif line.startswith("- Lunch"):
            formatted += f"- Lunch (₹200)\n"
        elif line.startswith("Justification:"):
            formatted += "\nJustification:\n"
            formatted += lines[lines.index(line) + 1] + "\n"
        elif line.startswith("Total Cost:"):
            formatted += f"\n💸 Total Cost: ₹{total_cost}\n"

    # Fallback: Add places if LLM didn't include enough
    for i in range(days):
        if not places_added[i]:  # No places added for this day
            formatted += f"🗓️ Day {i+1}:\n"
            places_to_add = [doc for doc in filtered_docs if doc.metadata['name'] not in sum(places_added, [])]
            for doc in places_to_add[:2]:  # Add up to 2 places
                fee = int(doc.metadata['entry_fee'])
                duration = float(doc.metadata['duration_hours'])
                if total_cost + fee <= budget and day_hours[i] + duration <= 8:
                    formatted += f"- {doc.metadata['name']} (₹{fee}, {duration} hours)\n"
                    total_cost += fee
                    day_hours[i] += duration
            formatted += f"- Lunch (₹200)\n"

    formatted += f"\n💸 Total Cost: ₹{total_cost}\n"
    return formatted

# 7. Take user input
def get_user_input():
    destination = input("Enter destination city (e.g., Ooty): ").strip()
    days = int(input("Enter number of days: ").strip())
    budget = int(input("Enter budget (INR): ").strip())
    user_query = input("Enter preferences (e.g., waterfalls and nature spots): ").strip()
    return destination, days, budget, user_query

# 8. Run the pipeline
if __name__ == "__main__":
    destination, days, budget, user_query = get_user_input()
    itinerary = generate_itinerary(destination, user_query, budget, days)
    print("\n" + itinerary)
```

```python
import json

def generate_text_directions(distance_matrix_path):
    """Generate text-based directions from a JSON distance matrix file."""
    with open(distance_matrix_path, 'r') as f:
        data = json.load(f)

    directions = []
    for source, destinations in data.items():
        for dest, time in destinations.items():
            directions.append(f"From {source} to {dest}: {time} mins by auto")
    return directions

# File paths to your distance JSON files
distance_files = {
    "Ooty": "ooty_distances.json",
    "Munnar": "munnar_distances.json",
    "Coorg": "coorg_distances.json",
    "Wayanad": "wayanad_distances.json"
}

# Generate and print sample directions for each city
for city, file_path in distance_files.items():
    print(f"\n📍 {city} Directions:")
    directions = generate_text_directions(file_path)
    for line in directions[:5]:  # Show only first 5 routes for demo
        print(line)
```

```python
# 🔧 Step 1: Install ipywidgets if not already
!pip install ipywidgets

# 📦 Step 2: Import necessary packages
import json
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

# 📂 Step 3: Load the places JSON data
with open("places.json", "r") as f:
    places_data = json.load(f)

# 🔍 Step 4: Extract city and type info for filtering
cities = sorted(list(set([place["city"] for place in places_data])))
types = sorted(list(set([place["type"] for place in places_data])))

# 🧩 Step 5: Define widgets
city_dropdown = widgets.Dropdown(options=cities, description="City:")
type_dropdown = widgets.Dropdown(options=["any"] + types, description="Type:")
days_slider = widgets.IntSlider(value=1, min=1, max=7, step=1, description="Days:")
button = widgets.Button(description="Generate Itinerary")

# 📜 Step 6: Define itinerary generation logic
def generate_itinerary(city, travel_type, days):
    filtered = [p for p in places_data if p["city"] == city]
    if travel_type != "any":
        filtered = [p for p in filtered if p["type"] == travel_type]

    random.shuffle(filtered)
    itinerary = filtered[:days * 2]  # 2 places per day
    return itinerary

# 🚀 Step 7: Button click logic
def on_button_click(b):
    clear_output(wait=True)
    display(city_dropdown, type_dropdown, days_slider, button)

    city = city_dropdown.value
    ttype = type_dropdown.value
    days = days_slider.value

    result = generate_itinerary(city, ttype, days)
    print(f"\n🎯 Suggested {days}-Day Itinerary for {city} ({ttype if ttype != 'any' else 'mixed'} type):\n")
    for i, place in enumerate(result, 1):
        print(f"{i}. {place['name']} - {place['type']} ({place['duration_hours']} hrs) - ₹{place['entry_fee']} - ⭐ {place['rating']}")
        print(f"   📍 {place['description']}\n")

# 🔗 Step 8: Attach event
button.on_click(on_button_click)

# 🎯 Step 9: Display UI
display(city_dropdown, type_dropdown, days_slider, button)
```
